```python
import cv2
import rasterio
import matplotlib.pyplot as plt
import numpy as np
%matplotlib tk
```

```python
# .tif画像を読み込む
with rasterio.open("/home/aichi2204/Documents/blk2go/20240412-library2/aichi-20240412-library2_CSF_flat_100mm_500_500mm/aichi-20240412-library2_CSF_flat_100mm_500_500mm.tif") as src:
    # 全バンドを読み込む
    image = src.read()
```

```python
print(f"Image shape: {image.shape}")  # (バンド数, 高さ, 幅)
```

```python
# 画像を表示
plt.imshow(image[4], cmap='gray')  # グレースケール表示
plt.colorbar()  # カラーバーを追加
plt.title("TIF Image")
plt.show()
```

```python
image_rgb = np.stack((image[0],image[1],image[2]), axis=-1).astype(int)

plt.imshow(image_rgb)
plt.title("RGB TIF Image")
plt.show()
```
